**Ventas totales supermercado**

```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .csv sin indice
data = pd.read_csv('Ventas_totales_supermercados.csv')
data.head(5)
```

```python
data.info()
```

**Todos los registros de “indice_tiempo: posteriores a 2017-02-01**

```python
# Filtrar registros donde ""indice_tiempo"" es posterior a 2017-02-01
filtro1=data[(data["indice_tiempo"] >"2017-02-01")]
filtro1
```

```python
# Filtrar registros donde "ventas_precios_corrientes" es mayor que 50,000 y menor que 120,000
filtro2 = data[(data["ventas_precios_corrientes"] > 50000) & (data["ventas_precios_corrientes"] < 120000)]

# Mostrar el resultado del filtro
filtro2
```

```python
#Todos los registros de “ ventas_precios_constantes : menor que 25, 000 o mayor que 27,000
# Filtrar registros donde "ventas_precios_constantes" es menor que 25,000 o mayor que 27,000
filtro3 = data[(data["ventas_precios_constantes"] < 25000) | (data["ventas_precios_constantes"] > 27000)]

# Mostrar el resultado del filtro
filtro3
```

```python
#Todos los registros con “ventas_totales_canal_venta” mayor a 2.796900e+07 y con “salon_ventas” menor a 1.643211e+08
# Filtrar registros donde "ventas_totales_canal_venta" es mayor que 2.796900e+07 y "salon_ventas" es menor que 1.643211e+08
filtro4 = data[(data["ventas_totales_canal_venta"] > 2.796900e+07) & (data["salon_ventas"] < 1.643211e+08)]

# Mostrar el resultado del filtro
filtro4
```

```python
#Todos los registros con canales_on_line: igual a 3.996933e+05 o ventas_totales_medio_pago” mayor a 2.913427e+07
# Filtrar registros donde "canales_on_line" es igual a 3.996933e+05 o "ventas_totales_medio_pago" es mayor que 2.913427e+07
filtro5 = data[(data["canales_on_line"] == 3.996933e+05) | (data["ventas_totales_medio_pago"] > 2.913427e+07)]

# Mostrar el resultado del filtro
filtro5
```

```python
#Todos los registros que no sean de efectivo : mayor que 696260e+07
# Filtrar registros donde "efectivo" no sea mayor que 6.96260e+07
filtro6 = data[~(data["efectivo"] > 6.96260e+07)]

# Mostrar el resultado del filtro
filtro6
```

```python
#Todos los registros que no sean de “verduleria_fruteria“ mayor a 6.138441e+06 o menor a 9.8e+05
# Filtrar registros donde "verduleria_fruteria" no sea mayor que 6.138441e+06 o menor que 9.8e+05
filtro7 = data[~((data["verduleria_fruteria"] > 6.138441e+06) | (data["verduleria_fruteria"] < 9.8e+05))]

# Mostrar el resultado del filtro
filtro7
```
